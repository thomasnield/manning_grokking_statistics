```python
import random

# Number of trials
n_trials=1_000_000

# Number of successful trials
succeeded = 0

# Simulate the lady tasting tea
for _ in range(n_trials):

    # 8 cups labeled 0-7
    cups = list(range(8))

    # Randomly select 4 for milk first (actual)
    true_milk_first = set(random.sample(cups, 4))

    # Lady's random guess on 4 cups with milk first
    guess_milk_first = set(random.sample(cups, 4))

    # Number correct: intersection size
    correct = len(true_milk_first & guess_milk_first)
    if correct == 4:
        succeeded += 1

print(f"\nProportion of outcomes with all 4 correct: {succeeded / n_trials:.4f}")
```

```python
from math import sqrt
from scipy.stats import norm

LEFT_TAIL, RIGHT_TAIL, TWO_TAIL = 0, 1, 2

# MODIFY PARAMETERS IN THIS BOX
# ==============================
mu = 142.0
n = 41 # must be greater than 30
x_bar = 140.2
s = 4.79
alpha = 0.05
tail_type = LEFT_TAIL
# ==============================

# Calculate z-statistic
z_stat = (x_bar - mu) / (s / sqrt(n))

# Initialize our pending result
reject_H0 = None

# Calculate critical value and compare it against z-statistic
if tail_type == LEFT_TAIL:
    critical_value = norm.ppf(alpha)
    reject_H0 = z_stat < critical_value

elif tail_type == RIGHT_TAIL:
    critical_value = norm.ppf(1 - alpha)
    reject_H0 = z_stat > critical_value

elif tail_type == TWO_TAIL:
    critical_value = norm.ppf(1 - (alpha / 2))
    reject_H0 = abs(z_stat) > critical_value

else:
     raise ValueError("Invalid tail type. Choose from LEFT_TAIL, RIGHT_TAIL, or TWO_TAIL.")

# Output results
if tail_type == LEFT_TAIL:
    print(f"LEFT TAIL TEST")
elif tail_type == RIGHT_TAIL:
    print(f"RIGHT TAIL TEST")
elif tail_type == TWO_TAIL:
    print(f"TWO-TAIL TEST")

print("Sample mean: ", x_bar)  #H
print("Sample standard deviation: ", s)  #H
print("Sample size: ", n)  #H
print("Level of confidence: ", 1 - alpha)  #H
print("Level of significance: ", alpha)  #H
print(f"z-statistic: {z_stat:.4f}")
print(f"Critical value: {critical_value:.4f}")

# Decision based on p-value
if reject_H0:
    print(f"Reject the null hypothesis at alpha = {alpha}. There is evidence to support the alternative hypothesis.")
else:
    print(f"Fail to reject the null hypothesis at alpha = {alpha}. There is not enough evidence to support the alternative hypothesis.")
```
